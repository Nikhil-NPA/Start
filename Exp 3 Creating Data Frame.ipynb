# Adding Records(Row) to the Data Frame

```python
import pandas as pd
```

```python
df2=pd.DataFrame([[15,17,29,21]],
columns=["CD","DBMS","DSS","CAO"])
```

```python
df3 = pd.concat([df, df2], ignore_index=True)
```

```python
df3
```

```python
df3.shape
```

```python
df3.size
```

# Adding Attribute (Columns) name DM

```python
df3["DM"]=[12,15,16,14]
```

```python
df3
```

# Deleting Record from df3 Data Frame

```python
df4=df3.drop(index=[1])
```

```python
df4
```

# Deleting Attribute (Columns) from df3 Data Frame

```python
df5=df3.drop(columns=["DM"])
```

```python
df5
```

```python
print("Mean of DSS:",df5["DSS"].mean())
```

```python
print("Median of DSS:",df5["DSS"].median())
```

```python
print("Mode of DSS:",df5["DSS"].mode())
```
